```python
import pandas as pd
```

```python
adata_train = pd.read_parquet('data/adata_train.parquet')
adata_train
```

```python
adata_train_pivoted = adata_train.pivot(index='obs_id', columns='gene', values='count')
adata_train_pivoted
```

```python
multiome_train = pd.read_parquet('data/multiome_train.parquet')
multiome_train
```

```python
multiome_train_pivoted = multiome_train.pivot(index='obs_id', columns='location', values='count')
multiome_train_pivoted
```

```python
import pickle

pickle.dump(multiome_train_pivoted, open('multiome_train_pivoted_raw_count.pkl', 'wb+'), protocol=5)
```
